# Visualization 
# 

```python
salaries = DS['BasePay']

plt.hist(salaries, bins=20, color='skyblue', edgecolor='black')

plt.xlabel('Salary')
plt.ylabel('Number of Employees')
plt.title('Distribution of Salaries')

plt.show()
```

```python
job_titles = DS['JobTitle']

department_counts = job_titles.value_counts()

plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90)

plt.title('Proportion of Employees in Different Departments')

plt.show()
```

```python
grouped_data = DS.groupby(['JobTitle', 'Year'])


average_salary = grouped_data['TotalPay'].mean()

print(average_salary)
```

```python
grouped = DS.groupby('JobTitle').agg({'TotalPay': ['mean', 'min', 'max', 'sum', 'count']})

print(grouped)
```

```python
grouped = DS.groupby('EmployeeName').agg({'TotalPay': ['mean', 'min', 'max', 'sum', 'count']})

print(grouped)
```

```python
salary = DS['TotalPay']
other_column = DS['BasePay']  

correlation_coefficient = salary.corr(other_column)

plt.scatter(other_column, salary)
plt.xlabel('Other Column')
plt.ylabel('Salary')
plt.title('Scatter Plot: Salary vs Other Column')
plt.show()

print('Correlation Coefficient:', correlation_coefficient)
```
